```python
import os
from pathlib import Path
import shutil
```

```python
path_eggs = 'folders_aux/'
AUDIO_FOLDER = 'Audios'
TEXT_FOLDER = 'Textos'
PATHS = ['audio_with_noise/', 'audios_denoised_dnn/', 'audios_denoised_wt/']
```

```python
folders_complement_noise_db = os.listdir(PATHS[0])
folders_complement_denoised_dnn_db = os.listdir(PATHS[1])
folders_complement_denoised_dnn_wt_db = os.listdir(PATHS[2])
```

```python
folders_complement_denoised_dnn_wt_db
```

```python
#As each folder share the same directories, so only make a list with all folders
ALL_FOLDERS_NOISE = []
for db in folders_complement_noise_db:
    for folders_noise in os.listdir(os.path.join(PATHS[0], db)):
        ALL_FOLDERS_NOISE.append(os.path.join(PATHS[0], db, folders_noise))
ALL_FOLDERS_NOISE.sort()
```

```python
ALL_FOLDERS_NOISY_BASELINE = []
ALL_FOLDERS_DNN_BASELINE = []
ALL_FOLDERS_TW_BASELINE = []

ALL_FOLDERS_NOISY_AUDIO_BASELINE = []
ALL_FOLDERS_DNN_AUDIO_BASELINE = []
ALL_FOLDERS_TW_AUDIO_BASELINE = []

ALL_FOLDERS_NOISY_TEXT_BASELINE = []
ALL_FOLDERS_DNN_TEXT_BASELINE = []
ALL_FOLDERS_TW_TEXT_BASELINE = []

for make_folder in ALL_FOLDERS_NOISE:
    for main_folder in PATHS:
        if(main_folder=='audio_with_noise/'):
            #print('1')
            ALL_FOLDERS_NOISY_AUDIO_BASELINE.append(os.path.join(path_eggs, main_folder, "/".join(make_folder.split('/')[1:len(make_folder)]), AUDIO_FOLDER))
            ALL_FOLDERS_NOISY_TEXT_BASELINE.append(os.path.join(path_eggs, main_folder, "/".join(make_folder.split('/')[1:len(make_folder)]), TEXT_FOLDER))
            ALL_FOLDERS_NOISY_BASELINE.append(os.path.join(main_folder, "/".join(make_folder.split('/')[1:len(make_folder)])))
            #print(os.path.join(main_folder, "/".join(make_folder.split('/')[1:len(make_folder)])))
        if(main_folder=='audios_denoised_dnn/'):
            #print('2')
            ALL_FOLDERS_DNN_AUDIO_BASELINE.append(os.path.join(path_eggs, main_folder, "/".join(make_folder.split('/')[1:len(make_folder)]), AUDIO_FOLDER))
            ALL_FOLDERS_DNN_TEXT_BASELINE.append(os.path.join(path_eggs, main_folder, "/".join(make_folder.split('/')[1:len(make_folder)]), TEXT_FOLDER))
            ALL_FOLDERS_DNN_BASELINE.append(os.path.join(main_folder, "/".join(make_folder.split('/')[1:len(make_folder)])))
            #print(os.path.join(main_folder, "/".join(make_folder.split('/')[1:len(make_folder)])))
        if(main_folder=='audios_denoised_wt/'):
            #print('3')
            ALL_FOLDERS_TW_AUDIO_BASELINE.append(os.path.join(path_eggs, main_folder, "/".join(make_folder.split('/')[1:len(make_folder)]), AUDIO_FOLDER))
            ALL_FOLDERS_TW_TEXT_BASELINE.append(os.path.join(path_eggs, main_folder, "/".join(make_folder.split('/')[1:len(make_folder)]), TEXT_FOLDER))
            ALL_FOLDERS_TW_BASELINE.append(os.path.join(main_folder, "/".join(make_folder.split('/')[1:len(make_folder)])))
            #print(os.path.join(main_folder, "/".join(make_folder.split('/')[1:len(make_folder)])))
        
        Path(os.path.join(path_eggs, main_folder, "/".join(make_folder.split('/')[1:len(make_folder)]), AUDIO_FOLDER)).mkdir(parents=True, exist_ok=True)
        Path(os.path.join(path_eggs, main_folder, "/".join(make_folder.split('/')[1:len(make_folder)]), TEXT_FOLDER)).mkdir(parents=True, exist_ok=True)
```

```python
path_texts = 'folders_aux/audio_clean/Textos/'
TEXTS = []
for t in os.listdir(path_texts):
    TEXTS.append(os.path.join(path_texts, t))
```

```python
def generate_data_asr(PATH1, PATH2, PATH3, PATH_TEXTS):
    for idx, (p1, p2) in enumerate(zip(PATH1, PATH2)):
        #print(p1)
        #print(p2)
        for audio in os.listdir(p1):
            shutil.copy(os.path.join(p1, audio), p2)
            #print(os.path.join(p1, audio))
        #print(p)
        #print(os.listdir(p))
        #break"""
    for p3 in PATH3: 
        #print(p3)
        for t in PATH_TEXTS:
            shutil.copy(t, p3)
            #print(t)
            #break
        #print(t)
```
