```python
import warnings
import pandas as pd
import numpy as np
import os
import pickle
import random
warnings.filterwarnings(action='ignore')
```

```python
def create_dataframe_analysis(root_folder):
    COUNTS = []
    for folder_name in os.listdir(root_folder):
        folder_path = os.path.join(root_folder, folder_name)
        if os.path.isdir(folder_path):
            count = 0
            for filename in os.listdir(folder_path):
                file_path = os.path.join(folder_path, filename)
                if os.path.isfile(file_path):
                    if filename.endswith(".flac"):
                        count = count + 1
            COUNTS.append((folder_name, count))
    return COUNTS
```

```python
def qualifying_speakers(folder_root, no_audios_min):
    COUNTS_SPEAKERS = create_dataframe_analysis(folder_root)
    #print(COUNTS_SPEAKERS)
    DF_SPEAKER = pd.DataFrame(COUNTS_SPEAKERS, columns=['Speaker', 'No_Audios'])
    #print(DF_SPEAKER)
    MINIMO_SPEAKER = DF_SPEAKER['No_Audios'].min()
    #print(MINIMO_SPEAKER)
    AUDIOS_MIN = DF_SPEAKER.nsmallest(5, 'No_Audios')
    print(AUDIOS_MIN)
    COUNT_SPEAKER = DF_SPEAKER['Speaker'].count()
    print('Speakers -> {:0}'.format(COUNT_SPEAKER))
    for index, row in DF_SPEAKER.iterrows():
        if(row['No_Audios']<no_audios_min):
            DF_SPEAKER = DF_SPEAKER.set_index('Speaker')
            DF_SPEAKER = DF_SPEAKER.drop(row['Speaker']) #This speaker has only 26 audios, so we drop this speaker
            DF_SPEAKER = DF_SPEAKER.reset_index()
        #else:
            #print(row['Speaker'])
        #print(row['Speaker'], row['No_Audios'])
    return DF_SPEAKER
```

```python
folder_root_sr = 'data/Speakers/train-clean-100/'
no_audios_min_sr = 100
DF_SR = qualifying_speakers(folder_root_sr, no_audios_min_sr)
print(DF_SR.nsmallest(20, 'No_Audios'))
COUNT_SPEAKER_SR = DF_SR['Speaker'].count()
print('Speakers -> {:0}'.format(COUNT_SPEAKER_SR))
```

```python
def read_a_certain_amount_of_audios(root_folder, amount_of_audios, amount_of_speakers, allow_audios):
    ID_SPEAKER = []
    ROOT_MAIN = []
    FILES = []
    countSpeaker = 0
    folder = os.listdir(root_folder)
    folder.sort()
    for folder_name in folder:
        folder_path = os.path.join(root_folder, folder_name)
        print(folder_path)
        #print(folder_path.split('/')[3])
        #Aqui ojo con el indice [4] y los paths que se estan trabajando
        if(folder_path.split('/')[3] in allow_audios):
            ROOT_MAIN.append(folder_path)
            ID_SPEAKER.append(folder_path.split('/')[3])
            if os.path.isdir(folder_path):
                files_audio = os.listdir(folder_path)
                lista_wav = [os.path.join(folder_path, archivo) for archivo in files_audio if archivo.endswith(".flac")]
                files_audio = lista_wav
                files_audio.sort()
                print(files_audio[0:amount_of_audios])
                FILES.append(files_audio[0:amount_of_audios])
                #for filename in files_audio[0:amount_of_audios]:
                #    print(os.path.join(folder_path, filename))
            countSpeaker = countSpeaker + 1
            if(countSpeaker>=amount_of_speakers):
                break
        else:
            print('No pass')
    return ID_SPEAKER, ROOT_MAIN, FILES
```

```python
AUDIO_PER_SPEAKER = 100
TOTAL_CLASSES = 20
```

```python
# SE EJECUTA
ROOT_FOLDER_CLEAN_AUDIOS = 'data/Speakers/train-clean-100/'
AMOUNT_OF_AUDIOS_CLEAN = AUDIO_PER_SPEAKER
AMOUNT_OF_SPEAKERS_CLEAN = TOTAL_CLASSES
ID_SPEAKER_CLEAN, ROOT_MAIN_CLEAN, FILES_CLEAN = read_a_certain_amount_of_audios(ROOT_FOLDER_CLEAN_AUDIOS, AMOUNT_OF_AUDIOS_CLEAN, AMOUNT_OF_SPEAKERS_CLEAN, DF_SR['Speaker'].values)
```
