```python
def get_clean_noise_data_load_from_pt_file(clean_files, noise_files):
    clean_data = []
    noise_data = []
    idx = []
    for clean_file, noise_file in zip(clean_files, noise_files):
        clean_data.append(torch.load(clean_file[1]))
        noise_data.append(torch.load(noise_file[1]))
        idx.append(clean_file[0])
    return idx, clean_data, noise_data
```

```python
class CustomDataset(Dataset):
    def __init__(self, idx, clean_data, noise_data):
        self.idx = idx
        self.clean_data = clean_data
        self.noise_data = noise_data

    def __len__(self):
        return len(self.clean_data)

    def __getitem__(self, index):
        idx = int(self.idx[index])
        clean_data = self.clean_data[index]
        noise_data = self.noise_data[index]
        return idx, clean_data, noise_data
```
